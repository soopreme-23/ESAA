# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

```python
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv',sep=',')
```

### Step 3. Assign it to a variable apple

```python
apple = data
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
apple['Date'] = pd.to_datetime(apple['Date'])
apple.dtypes
```

### Step 6.  Set the date as the index

```python
if 'Date' not in apple.columns :
  apple = apple.reset_index()

apple.set_index('Date',inplace=True)
```

### Step 7.  Is there any duplicate dates?

```python
# 중복된 날짜 있는 지 확인하기
duplicated_exist = apple.index.duplicated().any()
print('Duplicated_exist? :', duplicated_exist)

# 중복된 날짜가 있다면, 중복된 날짜 출력하기
duplicate_dates = apple.index[apple.index.duplicated()]
print('Duplicated dates : ', duplicate_dates)
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple.sort_index(inplace=True)
apple
```

### Step 9. Get the last business day of each month

```python
last_business_days = apple.resample('BM').last()
print(last_business_days)
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
days_diff = (apple.index.max() - apple.index.min()).days
print('Difference in days: ', days_diff)
```

### Step 11.  How many months in the data we have?

```python
n_months = apple.resample('ME').size().shape[0]
print(n_months)
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13.5,9))
apple['Adj Close'].plot()
plt.title('Apple Adjusted Close Price')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.grid(True)
plt.show()
```
