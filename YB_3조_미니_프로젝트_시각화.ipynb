# **2-1. 상영등급별 평균 관객수 분석 결과**

## **결론 :**

### **'청소년 관람불가' 등급이 평균 관객 수가 가장 높음**
- 일반적으로 청소년 관람불가 등급 영화는 성인 대상의 스릴러, 범죄, 느와르 장르가 많아 몰입도가 높고 충성도 있는 관객층을 형성할 수 있는 것으로 보임. 이로 인해 상대적으로 높은 관객 수를 기록했을 것으로 추론할 수 있음

### **'전체 관람가' 등급 영화는 평균 관객수가 가장 낮음**
- 가족/어린이 대상 콘텐츠가 가장 관객수가 낮았는데, 이는 마케팅이나 주제 측면에서 제한적이거나, 흥행성과는 큰 연관이 없을 수 있음. 이에 대한 분석은 추가적으로 필요해 보임

### **'15세 관람가'와 '12세 관람가'는 중간 수준**
- 대중성과 접근성의 균형을 이룬 등급으로, 전 연령대 타겟 영화가 많이 포함

# **2-2. 상영등급별 평균 관객수 분석 결과 : 비율 중심**

##**결론 :**

### **'청소년 관람불가' 등급의 흥행 비율이 가장 높음**
-  이 등급의 영화가 평균 관객수를 넘기는 경우가 많다는 의미로, 흥행 가능성이 상대적으로 크다는 것을 시사함

### **'15세 관람가'와 '12세 관람가' 영화는 중간 수준**
- 특정 연령 타겟을 갖는 만큼 안정적인 성적을 낼 수 있는 카테고리임을 보여줌

### **'전체 관람가' 등급의 흥행 비율이 가장 낮음**
- 가족 영화나 어린이 대상 영화의 경우 흥행하기 어려운 경향이 있음을 알 수 있음


```python
average_audience = data['box_off_num'].mean()
data['is_hit'] = data['box_off_num'] >= average_audience
rating_hit_ratio = data.groupby('screening_rat')['is_hit'].mean().sort_values(ascending=False)

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'AppleGothic'

plt.figure(figsize=(10,6))
sns.barplot(x=rating_hit_ratio.index, y=rating_hit_ratio.values, palette = 'viridis')
plt.title('rating_hit_ratio')
plt.xlabel('level')
plt.ylabel('ratio')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **3. 장르와 상영등급별 흥행 비율 히트맵 분석**

## **결론 :**

### **느와르 + 청소년 관람불가 조합 : 흥행 비율 높음**
- 성인 타깃이 강렬한 장르가 확실한 관객층을 보유하고 있음을 나타냄

### **액션, 드라마, 범죄 장르 : 일반적으로 15세 또는 청소년 관람불가 등급일 때 흥행비율이 높음.**
- 이는 영화의 몰입도와 강렬한 테마가 성인 타깃에게 잘 작동함을 보여줌

### **코미디 + 전체관람가 / 12세 관람가 : 흥행 비율 낮음.**
- 가볍고 넣은 타깃층을 노리는 전략이 오히려 관객 집중도를 낮출 수 있음을 보여줌


### **애니메이션 장르 : 전반적으로 흥행률이 낮음**
- 대중적 장르이지만 평균 관객수 기준으로는 흥행 성공률이 낮다는 점에서 마케팅/타겟 설정의 개선 여지 존재
- 하지만 매니아층이 있는 분야인 만큼 흥행을 책정하는 기준이 다르므로 이에 대한 추가적인 고려가 필요해보임

```python
pivot = pd.pivot_table(data, values='is_hit',index='genre', columns='screening_rat', aggfunc='mean').fillna(0)

plt.figure(figsize=(12,8))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('rating_hit_ratio_by_genre')
plt.xlabel('screening_rat')
plt.ylabel('genre')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/movies_train.csv')
```

```python
def get_season(release_time):
  # release_time에서 월을 추출
  month = release_time.month

  # 각 월에 맞는 분기를 결정
  if month in [1,2,3]:
    return 'first'
  elif month in [4,5,6]:
    return 'second'
  elif month in [7,8,9]:
    return 'third'
  elif month in [10,11,12]:
    return 'fourth'
```

```python
# release_time을 datetime 형식으로 변환
data['release_time'] = pd.to_datetime(data['release_time'])

# 월을 추출하여 새로운 컬럼 'month' 추가
data['month'] = data['release_time'].dt.month

# 월별 관객 수 합계 계산
monthly_box_off = data.groupby('month')['box_off_num'].sum().reset_index()

# 월별 관객 수 시각화 (막대그래프)
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='box_off_num', data=monthly_box_off, palette='viridis')
plt.title('Monthly Box_off_num')
plt.xlabel('month')
plt.ylabel('box_off_num')
plt.xticks(ticks=range(12), labels=['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'])
plt.show()
```

### **4-1.12/1월의 관객 수 (겨울 방학)**

- **12/1월 모두 관객 수 1,2위 장르는 드라마 아니면 코미디인데 관객수는 확연히 다름**

```python
# 12월 데이터만 필터링
data_december = data[data['release_time'].dt.month == 12]

# 장르별로 관객 수 합산
genre_box_off_december = data_december.groupby('genre')['box_off_num'].sum().reset_index()

# 관객 수를 기준으로 내림차순 정렬하고, 상위 5개 장르 선택
top_5_genres = genre_box_off_december.sort_values(by='box_off_num', ascending=False).head(5)

# 결과 출력
print("12월 중 가장 많은 관객을 기록한 상위 5개 장르:")
print(top_5_genres)
```

### **+ 4-3. 12세, 15세 관람가 내 상영 등급 별 장르 수**

- 대부분 15세 관람가인 영화들이 많은 편. 일부는 12세 관람가가 더 많음

```python
# 'screening_rat'이 '12세 관람가' 또는 '15세 관람가'인 데이터 필터링
filtered_data = data[data['screening_rat'].isin(['12세 관람가', '15세 관람가'])]

# 각 상영 등급별로 장르 수 카운트
genre_counts = filtered_data.groupby(['screening_rat','genre']).size().reset_index(name='count')

# 시각화: 상영 등급별 장르 수를 막대 그래프로 표시
plt.figure(figsize=(10,6))
sns.barplot(x='genre', y='count', hue='screening_rat', data=genre_counts)
plt.title('Genre Count by Screening_rat', fontsize=16)
plt.xlabel('genre', fontsize=12)
plt.ylabel('# of movies', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- 12세 관람가는 주로 자극적이지 않은 다큐멘터리, 애니메이션 장르에서 15세 관람가보다 많은 영화 수를 가짐

```python
# 12세 관람가와 15세 관람가로 데이터를 필터링
count_12 = genre_counts[genre_counts['screening_rat'] == '12세 관람가']
count_15 = genre_counts[genre_counts['screening_rat'] == '15세 관람가']

# 두 데이터를 merge하여 한 행에 12세와 15세 영화 수를 비교
merged_counts = pd.merge(count_12, count_15, on='genre', suffixes=('_12', '_15'))

# 12세 영화 수가 15세 영화 수보다 더 많은 장르를 필터링
more_12_than_15 = merged_counts[merged_counts['count_12'] > merged_counts['count_15']]

# 결과 출력
print(more_12_than_15[['genre', 'count_12', 'count_15']])
```

# **6. 감독의 이전 영화 평균 관객 수 vs. 현재 영화 관객 수**

- 이상치 제거 (상위 5% 제거 기준)


```python
# 95% 분위수 값 계산
threshold = data['dir_prev_bfnum'].quantile(0.95)

# 이상치 제거
filtered_df = data[data['dir_prev_bfnum'] <= threshold]
```

```python
def remove_outliers(df, column):
    """IQR 방법을 이용해 이상치를 제거하는 함수"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 이상치를 제거할 컬럼 리스트
outlier_columns = ["num_staff", "num_actor", "box_off_num"]

# 각 컬럼에 대해 이상치 제거 적용
for col in outlier_columns:
    data = remove_outliers(data, col)

# 이상치 제거 후 데이터 크기 확인
print("이상치 제거 후 데이터 크기:", data.shape)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_df, x='dir_prev_bfnum', y='box_off_num', alpha=0.6)
plt.title('dir_prev_bfnum vs box_off_num (이상치 제거)')
plt.xlabel('dir_prev_bfnum')
plt.ylabel('box_off_num')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = data[['dir_prev_bfnum', 'box_off_num']].dropna()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='dir_prev_bfnum', y='box_off_num', alpha=0.6)
plt.title('dir_prev_bfnum vs box_off_num')
plt.xlabel('dir_prev_bfnum')
plt.ylabel('box_off_num')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot( # 회귀선 추가
    data=filtered_df,
    x='dir_prev_bfnum',
    y='box_off_num',
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'red'},
    ci=None # 신뢰구간 제거
)
plt.title('dir_prev_bfnum vs box_off_num (이상치 제거 + 회귀선)')
plt.xlabel('dir_prev_bfnum')
plt.ylabel('box_off_num')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## **결론 :**
### **양의 상관관계 존재**

* 회귀선이 기울기가 작은 우상향으로 나타나므로, 두 변수가 약한 양의 상관관계를 가진다는 것을 알 수 있음.

→ 과거 관객 수(dir_prev_bfnum)가 많을수록 이번 영화의 관객 수도 많아지는 경향이 있음.

### **산점도가 넓게 퍼져 있음**
* 과거 관객 수(dir_prev_bfnum)가 적었어도 이번에는 흥행에 성공한 영화도 있고,
과거 관객수가 많았어도 이번에는 흥행하지 못한 경우도 있음.

→ 따라서 절대적인 기준은 아님.

#### **즉, 감독의 과거 관객 수가 많을수록 흥행 가능성은 올라가지만, 다른 요인들도 함께 복합적으로 작용했다고 볼 수 있음**
